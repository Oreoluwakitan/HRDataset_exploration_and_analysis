```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import streamlit as st
```

```python
df = pd.read_csv('HRDataset_v14 org excel - Copy.csv')
```

```python
to_cast = ['DOB', 'DateofHire', 'DateofTermination']
for cols in to_cast:
    df[cols] = df[cols].astype('datetime64[ns]')
```

```python
df['TermReason']= df['TermReason'].str.replace('N/A-', '')
```

```python
df['Employee_Name']= df['Employee_Name'].str.replace(',', '')
```

```python
df = df.drop(columns = ['MarriedID', 'GenderID', 'FromDiversityJobFairID', 'Termd', 'LastPerformanceReview_Date'], axis = 0)
```
